```python
#Operadores comparação

print(5 < 3)
print(5 > 3)
print(5 <= 3)
print(5 >= 3)
print(5 == 3)
```

```python
#Operadores condicionais: São operadores que comparam dois valores e retornam um valor booleano: True (verdadeiro) ou False (falso).

if 5 > 3:
    print("5 é maior que 3")
else:   
    print("Não é")

```
