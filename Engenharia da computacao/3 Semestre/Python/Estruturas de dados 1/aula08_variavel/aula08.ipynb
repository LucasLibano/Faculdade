Para que um programa armazene informações, é necessário declarar variáveis. Uma variável é um nome que referencia um espaço reservado na memória RAM. Quando declaramos uma variável, o sistema reserva uma quantidade específica de bytes, dependendo do tipo de dado. Quando o valor é atribuído ou manipulado, o processador grava os bits correspondentes naquele endereço de memória usando sinais elétricos (pulsos elétricos).

```python
numero = 3
numero_jogo = 14
numeros_convidados = 50

print(numero,numero_jogo,numeros_convidados)
```
