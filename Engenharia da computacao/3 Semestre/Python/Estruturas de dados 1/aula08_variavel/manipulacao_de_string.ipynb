Uma string é simplesmente um texto, ou seja, uma sequência de letras, números ou símbolos, como "Olá", "123" ou "Python!".

Manipular strings é mexer no texto sem mudar o original: pegar letras (texto[0]), pegar partes (texto[0:3]), checar palavras ("Py" in texto), trocar palavras (texto.replace()), mudar maiúsculas/minúsculas (.upper()/ .lower()), dividir ou juntar (.split()/ .join()), sempre criando novas strings.

Temos o Python Core, mas com o passar do tempo vamos ter que usar bibliotecas do python para os meus projetos.

```python
#String
a = 'casaco'
print(a)

#Maiscula.
maiuscula = a.upper() #.upper É uma funcionalidade que pega uma varivavel e manipula ela o deixando maiuscula.
print(maiuscula)

#Minuscula
minuscula = a.lower() #.lower manipula a varivael para deixar minuscula
print(minuscula)
```

Todo caracter no python é uma ARRAY, então tem como retirar "pedaços". Toda array começa no numero 0 onde se vira a primeira "casa".

```python
a[0:4] # a[0:4] pega os primeiros 4 caracteres da string sem mudar a original
print(a)

ultimas_letras = a[3:] # ultimas_letras = a[3:] pega todos os caracteres a partir do índice 3 até o fim
print(ultimas_letras)

novo_texto = "XYZ"

subt = a.replace (ultimas_letras, novo_texto) # subt = a.replace(ultimas_letras, subt) cria uma nova string trocando ultimas_letras por subt
print(subt)

a.find('5') # a.find('5') procura o índice da primeira ocorrência de '5' (-1 se não encontrar)
a.find('6') # a.find('6') procura o índice da primeira ocorrência de '5' (-1 se não encontrar)

a = ' casaco ' 
print(len(a)) # len(a) retorna o tamanho da string, contando todos os caracteres incluindo espaços
```
